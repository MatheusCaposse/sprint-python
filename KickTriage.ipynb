```python
# Importa a função randrange do módulo random, que gera números aleatórios
from random import randrange

# Variável global que serve para contar o grau de gravidade dos sintomas
contador = 0

# Função que faz a triagem de sintomas relacionados à febre
def pergunta_febre():
    
    global contador  # Acessa a variável global contador
    
    # Pergunta se o paciente tem febre e converte a resposta para minúsculas
    tem_febre = str(input("O paciente apresenta sintomas de febre? ")).lower()

    temperatura = None  # Variável para armazenar a temperatura

    # Se o paciente tem febre, pede para informar a temperatura mais alta
    if "sim" in tem_febre:
        temperatura = int(input("Qual a temperatura mais alta que o paciente apresentou? "))
        # Se a temperatura for maior ou igual a 39, aumenta o contador em 2 (maior gravidade)
        if temperatura >= 39:
            contador+=2

    # Pergunta se o paciente tem pressão alta ou baixa e qual é o tipo
    pressao = str(input("O paciente apresente pressão alta ou baixa? se sim qual. "))
    
    # Retorna as respostas coletadas e o valor do contador
    return tem_febre, pressao, temperatura, contador

# Função que faz a triagem de outros sintomas
def pergunta_sintomas():
    global contador  # Acessa a variável global contador
    
    # Pergunta se o paciente tem dor em alguma parte do corpo
    dores = str(input("O paciente tem dor em alguma parte do corpo? ")).lower()

    local_dor = None  # Variável para armazenar o local da dor
    motivo_dor = None  # Variável para armazenar o motivo da dor
    intensidade_dor = None # Variável para armazenar a intensidade da dor

    # Se o paciente tem dor, pede para informar onde e o motivo
    if "sim" in dores:
        local_dor = str(input("Onde a dor se localiza? "))
        # Se a dor for no peito ou na cabeça, aumenta o contador em 2
        if local_dor == 'peito' or local_dor == 'cabeça':
            contador+=2

        motivo_dor = str(input("Qual o motivo da dor? (n/a caso não saiba)")).lower()
        # Se o motivo for fratura ou algo semelhante, aumenta o contador em 1
        if 'quebrada' in motivo_dor or 'fraturada' in motivo_dor:
            contador +=1
        
        # Pergunta a intensidade da dor
        intensidade_dor = int(input("De 0-10 a intensidade da dor, sendo 0 fraca e 10 muito forte"))
        # Adiciona pontos ao contador com base na intensidade da dor
        if 3 < intensidade_dor <= 6:
            contador += 1
        elif intensidade_dor >= 7:
            contador += 2
        
    # Pergunta se o paciente está vomitando, e aumenta o contador em 1 se a resposta for sim
    vomito = str(input("O paciente apresenta sintomas de vomito? "))
    if vomito == 'sim':
        contador += 1

    # Retorna todas as respostas e o valor atualizado do contador
    return dores, local_dor, motivo_dor, vomito, contador, intensidade_dor

# Função principal de triagem do paciente
def triagem():
    global contador  # Acessa a variável global contador 

    # Coleta dados básicos do paciente
    nome = str(input("Digite o seu nome completo: "))
    cpf = str(input("Digite seu cpf: "))
    idade = int(input("Digite a sua idade:"))

    # Chama a função de febre e armazena as respostas
    febre, pressao, temperatura, contador1 = pergunta_febre()

    # Chama a função de sintomas e armazena as respostas
    dor, local_dor, motivo_dor, vomito, contador2, intensidade_dor = pergunta_sintomas()

    # Calcula o total de pontos no contador
    contador_geral = contador1 + contador2

    # Define a gravidade com base no total de pontos
    gravidade = None
    if contador_geral <= 4:
        gravidade = 'Leve'
    elif 4 < contador_geral < 7:
        gravidade = 'Moderada'
    else:
        gravidade = 'Grave'

    # Pergunta quando os sintomas começaram
    data_sintomas = str(input("Quando esses sintomas começaram? "))

    # Dicionário que armazena as informações coletadas
    ficha_medica = {
        "Nome": nome,
        "Cpf": cpf,
        "Idade": idade,
        "Pressao": pressao, 
        "Vomitos": vomito,
        "Sintomas de febre": f"{febre}, {temperatura if 'sim' in febre else ''}",
        "Dores no corpo": f"{dor}, {local_dor and intensidade_dor if 'sim' in dor else ''}",
        "Quando começaram os sintomas": data_sintomas,
        "Motivor dor": motivo_dor if motivo_dor != 'n/a' else "N/A",
        "Gravidade": gravidade
    }

    # Retorna o dicionário com a ficha médica
    return ficha_medica

# Função para exibir as informações do paciente de forma formatada
def exibir_ficha(ficha):
    print("\n------- HOSPITAL SABARA -------\n")
    # Gera um número aleatório para a chamada do paciente
    numero_chamada = randrange(0,200)
    print(f"\n---- Numero de chamada: {numero_chamada} ----\n")
    # Exibe os dados do dicionário
    for chave, valor in ficha.items():
        print(f"{chave}: {valor}")
    print("-"*40)

# Chama a função triagem para coletar os dados do paciente
ficha = triagem()
# Exibe as informações do paciente
exibir_ficha(ficha)
```
